# Data Cleaning

```python
# Change data type that not adjust
#1.We can change date column from type object -> datetime

data['date'] = pd.to_datetime(data['date'])
data.info()
```

```python
# we look null in data and if > 80% we can delete that null
data.isnull().sum()
```

```python
# we can delete column recovered,active,fips,admin2
column = data[['recovered','active','fips','admin2']]
for i in column:
  del data[i]
```

```python
data.isnull().sum()
```

```python
# handle missing value
#1. fill province state that null with uknown

data['province_state'] = data['province_state'].fillna('Unknown')
data.isnull().sum()
```

```python
data.head(50)
```

```python
#2 we can delete rows that null from latitude,longitude,location_geom
#because if we look in data. Many data that null and irrelevant data
data = data.dropna()
```

```python
#adjust name column
data = data.rename(columns={'combined_key': 'Complete_address'})
```
